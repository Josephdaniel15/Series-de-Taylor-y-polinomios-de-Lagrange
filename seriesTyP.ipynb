```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-np.pi, np.pi, 256)
x0 = 1
y0 = np.cos(x0)

y = np.cos(x)
y1 = np.cos(x0) - np.sin(x0) * (x - x0)
y2 = y1 - (np.cos(x0) / 2) * (x - x0)**2
y3 = y2 + (np.sin(x0) / 6) * (x - x0)**3

plt.plot(x, y, label="cos(x)", color="black")
plt.plot(x, y1, label="Orden 1")
plt.plot(x, y2, label="Orden 2")
plt.plot(x, y3, label="Orden 3")
plt.axhline(y0, color="red", linestyle="--", label="Orden 0")

plt.xticks([-np.pi, 0, np.pi], [r"$-\pi$", "0", r"$\pi$"])
plt.ylim(-1.5, 1.5)
plt.legend()
plt.show()
print(f"p0 es {y0:.4f}")
print(f"p1 es {y0:.4f} - {np.sin(x0):.4f}(x - 1)")
print(f"p2 es {y0:.4f} - {np.sin(x0):.4f}(x - 1) - {(np.cos(x0)/2):.4f}(x - 1)^2")
print(f"p3 es {y0:.4f} - {np.sin(x0):.4f}(x - 1) - {(np.cos(x0)/2):.4f}(x - 1)^2 + {(np.sin(x0)/6):.4f}(x - 1)^3")
```

# Encuentre el polinomio de Lagrange para los siguientes datos y grafique:



```python
import matplotlib.pyplot as plt
import numpy as np

x_datos = [0, 30, 60, 90]
y_datos = [0, 0.5, np.sqrt(3)/2, 1]

def lagrange(x, xs, ys):
    p = 0

    for i in range(len(xs)):
        li = 1

        for j in range(len(xs)):
            if i != j:
                li *= (x - xs[j]) / (xs[i] - xs[j])

        p += ys[i] * li

    return p

x = np.linspace(0, 90, 300)

p1 = [lagrange(v, x_datos[:2], y_datos[:2]) for v in x]
p2 = [lagrange(v, x_datos[:3], y_datos[:3]) for v in x]
p3 = [lagrange(v, x_datos[:4], y_datos[:4]) for v in x]

plt.plot(x, p3, color="black", label="Polinomio Final")
plt.plot(x, p1, "--", label="Grado 1")
plt.plot(x, p2, "--", label="Grado 2")
plt.plot(x, p3, "--", label="Grado 3")

plt.scatter(x_datos, y_datos)

plt.legend()
plt.grid()
plt.show()

print("P1:")
print("0 + 0.5((x-0)/(30-0))")

print()
print("P2:")
print("0 + 0.5((x-0)(x-60))/((30-0)(30-60)) +")
print("0.8660((x-0)(x-30))/((60-0)(60-30))")

print()
print("P3:")
print("0 +")
print("0.5((x-0)(x-60)(x-90))/((30-0)(30-60)(30-90)) +")
print("0.8660((x-0)(x-30)(x-90))/((60-0)(60-30)(60-90)) +")
print("1((x-0)(x-30)(x-60))/((90-0)(90-30)(90-60))")
```
